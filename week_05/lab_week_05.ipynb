# Primary Research Question
How have the world record times for the men's and the womenâ€™s mile event changed over the years?

```r
table(wr$Event)
```

```r
mm = wr[wr$Event == 'Mens Mile',]
wm = wr[wr$Event == 'Womens Mile',]
```

```r
plot(mm$Year,mm$Record,main='mens mile , year vs record time',xlab='year',ylab='time')
plot(wm$Year,wm$Record,main='womens mile , year vs record time',xlab='year',ylab='time')
```

```r
linFit(mm$Year,mm$Record)
linFit(wm$Year,wm$Record)
```
