```python
import matplotlib.pylab as plt
%matplotlib inline
```

Graph the following exponential functions by making a table of values.
1. y = 3x
2. y = 5 * 3x
3. y = 40 * 4x
4. y = 3 * 10x

```python
y1 = lambda x : 3*x
```

```python
y2 = lambda x : 5 * 3 * x
```

```python
y3 = lambda x : 40 * 4 * x
```

```python
y4 = lambda x : 3 * 10 * x
```

### Graph y1

```python
x = range(-20,20)
```

```python
y = [y1(v) for v in x]
```

```python
print(y)
```

```python
fig,ax = plt.subplots(figsize=(15,5))
_ = ax.plot(x,y)
```
